```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread('alien.jpeg')

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
color = ('b', 'g', 'r')
for i, col in enumerate(color):
    histr = cv2.calcHist([img1], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
plt.title('Color Histogram')

plt.subplot(1, 2, 2)
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.title('Image')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt


img_bgr = cv2.imread('alien.jpeg', 1)
plt.imshow(img_bgr)
plt.axis('off')
plt.subplot(2, 2, 1)

color = ('b', 'g', 'r')

for i, col in enumerate(color):

	histr = cv2.calcHist([img_bgr],
						[i], None,
						[256],
						[0, 256])

	plt.plot(histr, color = col)

	plt.xlim([0, 256])
plt.subplot(2, 2, 2)

height, width, _ = img_bgr.shape

for i in range(0, height - 1):
	for j in range(0, width - 1):

		pixel = img_bgr[i, j]


		pixel[0] = 255 - pixel[0]

		pixel[1] = 255 - pixel[1]

		pixel[2] = 255 - pixel[2]

		img_bgr[i, j] = pixel

plt.imshow(img_bgr)
plt.subplot(2, 2, 3)

color = ('b', 'g', 'r')

for i, col in enumerate(color):

	histr = cv2.calcHist([img_bgr],
						[i], None,
						[256],
						[0, 256])

	plt.plot(histr, color = col)
	plt.xlim([0, 256])

plt.subplot(2, 2, 4)
```
